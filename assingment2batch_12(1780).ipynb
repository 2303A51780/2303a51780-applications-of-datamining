```python
#1780
import pandas as pd
url = '/content/netflix_titles.csv'
data = pd.read_csv(url)
summary = data.describe()
print(summary)
```

```python
import matplotlib.pyplot as plt
plt.hist(data['release_year'].dropna(), bins=20, edgecolor='black', color='brown')
plt.title('Release Distribution')
plt.xlabel('Release Year')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(6, 4))
data['type'].value_counts().plot(kind='bar', color='lightcoral')
plt.title('Content Type Distribution')
plt.xlabel('Type')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()
```

```python
plt.figure(figsize=(6, 4))
data['country'].value_counts().plot(kind='bar', color='lightcoral')
plt.title('Country Distribution')
plt.xlabel('Country')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(6, 4))
plt.boxplot(data['release_year'].dropna(), patch_artist=True,
boxprops=dict(facecolor='lightgreen'))
plt.title('Box Plot of Release Year')
plt.ylabel('Release Year')
plt.show()
```

```python
import seaborn as sns
sns.pairplot(data[['release_year', 'duration']].dropna())
plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
data['rating_encoded'] = le.fit_transform(data['rating'])
correlation_matrix = data[['release_year', 'rating_encoded']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', cbar=True)
plt.title('Correlation Heatmap')
plt.show()
```

```python
sns.countplot(x='duration', data=data, palette='Set2')
plt.title('duration Count')
plt.xlabel('duration')
plt.ylabel('Count')
plt.show()
```

```python
sns.violinplot(x='type', y='release_year', data=data, palette='muted')
plt.title('release_year Distribution by type')
plt.show()
```
