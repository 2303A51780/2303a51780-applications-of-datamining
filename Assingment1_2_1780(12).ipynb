```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
df[['HRA', 'DA']] = scaler.fit_transform(df[['HRA', 'DA']])
```

```python
df['HRA'] = (df['HRA'] - df['HRA'].min()) / (df['HRA'].max() - df['HRA'].min())
```
